```python
import pandas
from pandas import DataFrame
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data=pandas.read_csv('cost_revenue_clean.csv')
```

```python
data.describe()
```

```python
X=DataFrame(data,columns=['production_budget_usd'])
y=DataFrame(data,columns=['worldwide_gross_usd'])
```

```python
plt.figure(figsize=(10,6))
plt.title('film cost vs global revenue')
plt.xlabel('production budget $')
plt.ylabel('worldwide gross $')
plt.ylim(0,3000000000)
plt.xlim(0,450000000)
plt.scatter(X,y,alpha=0.3)
plt.show()
```

```python
regression=LinearRegression()
```

```python
regression.fit(X,y)
```

.coef_ = slope coefficient

```python
regression.coef_      # slope coefficient    theta_1
```

intrercept


```python
# intercept
regression.intercept_
```

plotting the line in the above graph using the matplotlib.pyplot module with the help of the plot

```python
plt.figure(figsize=(10,6))
plt.title('film cost vs global revenue')
plt.xlabel('production budget $')
plt.ylabel('worldwide gross $')
plt.plot(X,regression.predict(X), color='red',linewidth=4)
plt.ylim(0,3000000000)
plt.xlim(0,450000000)
plt.scatter(X,y,alpha=0.3)
plt.show()
```

to measure accuracy means by minimise  how to check

```python
regression.score(X,y)
```
